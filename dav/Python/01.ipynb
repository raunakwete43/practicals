```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('./Experience-Salary.csv')
df.head()
```

```python
print(df.isnull().sum())
print(sum(df.duplicated()))
```

```python
model = LinearRegression()
model.fit(df[['exp(in months)']], df['salary(in thousands)'])
predictions = model.predict(df[['exp(in months)']])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['exp(in months)'], df['salary(in thousands)'], color='blue', label='Actual Data')
plt.plot(df['exp(in months)'], predictions, color='red', linewidth=2, label='Regression Line')
plt.title('Experience vs Salary')
plt.xlabel('Experience (in months)')
plt.ylabel('Salary (in thousands)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print(f"Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")
```

```python
print(f"RMSE: {np.sqrt(mean_squared_error(df['salary(in thousands)'], predictions))}")
print(f"R^2 Score: {r2_score(df['salary(in thousands)'], predictions)}")
```
