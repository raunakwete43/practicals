```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'A': np.random.randn(100),
    'B': np.random.rand(100) * 100,
    'C': np.random.randint(0, 50, 100)
}
df = pd.DataFrame(data)

summary = df.describe()

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(df['A'], bins=20, color='skyblue')
plt.title('Histogram of A')

plt.subplot(1, 2, 2)
sns.scatterplot(x='B', y='C', data=df)
plt.title('Scatterplot of B vs C')
plt.tight_layout()
plt.show()

summary
```
